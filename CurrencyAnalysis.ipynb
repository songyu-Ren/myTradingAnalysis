```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
eur_cny = 'EURCNY=X'
usd_cny = 'USDCNY=X'
usd_eur = 'USDEUR=X'
```

```python
# Retrieve data for the last 4 years
start_date = '2019-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')  # Set end date to today
```

```python
# Fetch the data
eur_cny_data = yf.download(eur_cny, start=start_date, end=end_date)
usd_cny_data = yf.download(usd_cny, start=start_date, end=end_date)
usd_eur_data = yf.download(usd_eur, start=start_date, end=end_date)
```

```python
# Process and prepare data for plotting
eur_cny_data = eur_cny_data['Close']
usd_cny_data = usd_cny_data['Close']
usd_eur_data = usd_eur_data['Close']
# Create a DataFrame for combined data
combined_data = pd.DataFrame({'EUR/CNY': eur_cny_data, 'USD/CNY': usd_cny_data, 'USD/EUR': usd_eur_data})
# Drop rows with any missing values to ensure clean data
combined_data.dropna(inplace=True)
# Normalize data by setting the first value of each series to 1
normalized_data = combined_data / combined_data.iloc[0]

```

```python
# Plot the normalized data to visualize trends
plt.figure(figsize=(12, 6))
plt.plot(normalized_data.index, normalized_data['EUR/CNY'], label='EUR/CNY')
plt.plot(normalized_data.index, normalized_data['USD/CNY'], label='USD/CNY')
plt.plot(normalized_data.index, normalized_data['USD/EUR'], label='USD/EUR')
plt.xlabel('Date')
plt.ylabel('Normalized Exchange Rate')
plt.title('Normalized Trends of EUR/CNY, USD/CNY, and USD/EUR (Last 4 Years)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Retrieve data for the last 4 years
start_date_2024 = '2024-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')  # Set end date to today
# Fetch the data
eur_cny_data_2024 = yf.download(eur_cny, start=start_date_2024, end=end_date)['Close']
usd_cny_data_2024 = yf.download(usd_cny, start=start_date_2024, end=end_date)['Close']
usd_eur_data_2024 = yf.download(usd_eur, start=start_date_2024, end=end_date)['Close']
```

```python
# Combine data into a single DataFrame
combined_data_2024 = pd.DataFrame({
    'EUR/CNY': eur_cny_data_2024,
    'USD/CNY': usd_cny_data_2024,
    'USD/EUR': usd_eur_data_2024
})
# Drop rows with any missing values to ensure clean data
combined_data_2024.dropna(inplace=True)

# Normalize each series by setting its first value to 1
normalized_data_2024 = combined_data_2024 / combined_data_2024.iloc[0]

```

```python
# Plot the normalized data to visualize trends
plt.figure(figsize=(12, 6))
plt.plot(normalized_data_2024.index, normalized_data_2024['EUR/CNY'], label='EUR/CNY')
plt.plot(normalized_data_2024.index, normalized_data_2024['USD/CNY'], label='USD/CNY')
plt.plot(normalized_data_2024.index, normalized_data_2024['USD/EUR'], label='USD/EUR')
plt.xlabel('Date')
plt.ylabel('Normalized Exchange Rate')
plt.title('Normalized Trends of EUR/CNY, USD/CNY, and USD/EUR (2024)')
plt.legend()
plt.grid(True)
plt.show()
```
