```python
import os
import pandas as pd

# Get a list of all files in the current directory
files = [f for f in os.listdir('.') if os.path.isfile(f) and f.endswith('.csv')]

# Initialize an empty list to store DataFrames
dataframes = []

# Check if there are any CSV files
if files:
    # Loop through each CSV file found
    for file in files:
        df = pd.read_csv(file)
        dataframes.append(df)
    
    # Concatenate all DataFrames
    combined_df = pd.concat(dataframes)
    
    # Sort the combined DataFrame by 'Transaction Time (CET)'
    sorted_df = combined_df.sort_values(by='Transaction Time (CET)')
    
    # Get unique names from 'Asset Name' column
    unique_asset_names = sorted_df['Asset Name'].unique()
    
    # Print the unique asset names
    print("Unique Asset Names:")
    for name in unique_asset_names:
        print(name)
else:
    print("No CSV files found in the directory.")
```
